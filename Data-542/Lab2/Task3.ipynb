### create new columns as country and continent 

```python
t = '_Afghanistan'
a = t.split('_')
len(a[0])
```

```python
region = raw_df['region']
region.head(2)
```

```python
country = []
continent = []
for i,j in enumerate(region):
      t = j.split('_')
      # if t[1] == "Canada":
      #       print('t[0]',t[0])
      #       print('t[1]',t[1])
      if len(t[0]) > 0: 
            if len(t[1]) > 0: 
                  continent.append(t[0])
                  country.append(t[1])
            else:
                  continent.append('')
                  country.append(t[1])
      else:
            continent.append('')
            country.append(t[1])
```

```python
region_df = pd.DataFrame({'country': country, 'continent': continent})
region_df.head()
```

```python
raw_df = pd.concat([raw_df, region_df], axis=1)
raw_df = raw_df.drop(['region'], axis=1)
raw_df.head()
```
